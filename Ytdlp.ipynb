```

#@title üöÄ STEP 3: Download, Enhanced Clean & Merge
Format_ID = "136+140" #@param {type:"string"}

import os
import re
import glob
from google.colab import files

# 1. Download to a temporary file
print("--- [DEBUG] Downloading streams ---")
!yt-dlp --cookies cookies.txt \
        -f "{Format_ID}" \
        --merge-output-format mp4 \
        -o "TEMP_VAL.%(ext)s" \
        "{URL}"

# 2. Get the official Title (Now with cookies included!)
title_output = !yt-dlp --cookies cookies.txt --get-title "{URL}"

# Check if we actually got a title or an error
if title_output and "ERROR" not in title_output[0]:
    raw_title = title_output[0]
else:
    print("‚ö†Ô∏è Could not fetch title (Age Gate/Cookie issue). Using fallback name.")
    raw_title = "Video"

# 3. Enhanced Whitelist Filter
clean_title = re.sub(r'[^a-zA-Z0-9\s\!\.\?\,\;\:\(\)\[\]\-\_\'\"\@\#\$\%\&\+\=]', '', raw_title)
clean_title = re.sub(r'\s+', ' ', clean_title).strip()

# 4. Limit length to prevent "File name too long" error (Max 200 chars)
if len(clean_title) > 200:
    clean_title = clean_title[:200]

final_name = f"{clean_title} Youtube.mp4"

# 5. Rename and Trigger Download
if os.path.exists("TEMP_VAL.mp4"):
    if os.path.exists(final_name):
        os.remove(final_name)

    try:
        os.rename("TEMP_VAL.mp4", final_name)
        print(f"\n‚úÖ Cleaned Name: {final_name}")
        files.download(final_name)
    except OSError as e:
        print(f"‚ùå Still having naming issues: {e}")
        print("Downloading with fallback name instead.")
        files.download("TEMP_VAL.mp4")
else:
    print("\n‚ùå Error: MP4 file not found. Ensure Step 1 and 2 were successful.")
```
